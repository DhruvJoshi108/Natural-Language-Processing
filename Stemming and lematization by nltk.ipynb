```python
import nltk
```

```python
from nltk.stem import PorterStemmer
```

```python
ps=PorterStemmer()
```

"charging", "charges","charged","charge","study","studying","easily","fairly","losses","was","necessary","implementation","permission","learn"

```python
words=["charging", "charges","charged","charge","study","studying","easily","fairly","losses","was","necessary","implementation","permission","learn"]
```

```python
for word in words:
    print(word,":",ps.stem(word))
```

```python
sentence="Police is chasing the suspect as it ran away".split()
```

```python
for token in sentence:
    print(ps.stem(token))
```

## Snowball Stemmer

```python
import nltk
from nltk.stem.snowball import SnowballStemmer
snow_stemmer= SnowballStemmer(language='english')
```

```python
for word in words:
    print(snow_stemmer.stem(word))
```

```python
for word in sentence:
    print(snow_stemmer.stem(word))
```

Lemmatization with nltk

```python
from nltk.corpus import wordnet
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
lemmatizer.lemmatize("studying")
```

```python
lemmatizer.lemmatize("studying",pos=wordnet.VERB)
```
