## Date Format

```python
from datetime import datetime

today=datetime(year=2025, month=9,day=27)
print(f'{today:%B %d ,%Y}')
```

```python
today=datetime(year=2025, month=9,day=27)
print(f'{today:%b %d ,%Y}')
```

# Writing to a **File**

```python
# Add a second argument to the func,"w" which stands for write
#Passing "w+"
my_file=open('test.txt','w+')
```

```python
my_file.write('This is the new first line')
```

```python
my_file.close()
```

# Closing of files automatically using a context manager

```python
with open('test.txt','r') as txt:
  first_line=txt.readlines()[0]
print(first_line)


```

# Iterating through a file

```python
with open('test.txt','r') as txt:
  for line in txt:
    print(line,end='')
```

```python
with open('test.txt','r') as txt:
  for line in txt:
    print(line)
```
