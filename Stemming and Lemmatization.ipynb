**Import nltk and Porter Stemmer**

```python
import nltk 
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
w = ['charging' , 'charges' , 'charged' , 'charge' , 'study' , 'studying' , 'easily' , 'fairly' , 'losses' , 'was' , 'necessary' , 'implementation' , 'permission' , 'learn']
```

```python
for i in w:
    print(f"{i} : {ps.stem(i)}")
```
