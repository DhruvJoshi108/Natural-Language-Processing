 ***Apply NER to identify entities such as names, organizations, and locations in a given text. Perform sentence segmentation on a paragraph and explain its importance in NLP tasks.***



```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
para = '''
Data science has rapidly evolved into one of the most influential fields of the modern era, shaping decisions, innovations, and strategies across nearly every industry. At its core, data science is the practice of extracting meaningful insights from raw, often messy data. It blends several disciplines—statistics, computer science, mathematics, and domain expertise—to uncover patterns, make predictions, and support informed decision‑making. What makes data science so compelling is its ability to transform overwhelming volumes of information into clarity, enabling organizations and individuals to understand not just what is happening, but why it is happening and what is likely to happen next.

The data science process typically begins with data collection, which can involve anything from customer transactions and social media activity to sensor readings and medical records. Once collected, the data must be cleaned and prepared, a step that often takes more time than any other. Real‑world data is rarely perfect; it contains missing values, inconsistencies, and noise. Data scientists use a variety of techniques to organize, filter, and structure this information so it becomes suitable for analysis. This stage is crucial because the quality of insights depends heavily on the quality of the underlying data.

After preparation comes exploration and analysis. Data scientists use statistical methods and visualization tools to examine trends, correlations, and anomalies. This exploratory phase helps them form hypotheses and identify the most promising directions for deeper investigation. Machine learning—one of the most powerful tools in the data science toolkit—often plays a central role here. By training algorithms on historical data, data scientists can build models that classify information, detect patterns, or predict future outcomes. These models power everything from recommendation systems and fraud detection to medical diagnoses and autonomous vehicles.

However, data science is not just about algorithms and code. Communication is equally important. A data scientist must be able to translate complex findings into clear, actionable insights that non‑technical stakeholders can understand. This often involves creating visualizations, dashboards, or reports that highlight key results and explain their implications. The ability to tell a compelling story with data is what ultimately bridges the gap between analysis and real‑world impact.

Ethics also plays a growing role in data science. As organizations collect more data and build increasingly powerful models, concerns about privacy, fairness, and transparency have become central. Responsible data scientists must consider how their work affects individuals and society, ensuring that data is used ethically and that models do not reinforce harmful biases.

In essence, data science is both a technical discipline and a creative one. It requires curiosity, problem‑solving, and a willingness to explore the unknown. As data continues to grow in volume and importance, the role of data science will only expand, shaping the future of technology, business, and everyday life.
'''
```

```python
doc = nlp(para)
```

```python
print("Sentence segmentation:")
for sent in doc.sents:
    print("-", sent.text)
```

```python
print("\nNamed entities:")
for ent in doc.ents:
    print(ent.text, "->", ent.label_)
```
